```python
x_list = ['x1', 'x2']
y_list = ['y1', 'y2']

print("x y")
for x in x_list:
  for y in y_list:
    print(x, y)
```

```python
names = ['Jung', 'Lee', 'Song', 'Oh']
scores = [95, 96, 92, 94]

for k in range(len(names)): # len(names) = 4
  print(names[k], scores[k])
```

```python
for name, score in zip(names, scores):
  print(name, score)
```

### 반복문 while

반복 범위 없이 조건에 따라 반복

```python
i = 0
sum = 0

print("i sum")

while sum < 20:
  i = i + 1
  sum = sum + i
  print(i, sum)
```

```python
while True:
  print("while test")
```

반복문 제어하기

```python
k = 0
while True:
  k = k + 1

  if k > 3:
    break

  print(k)
```

```python
for k in range(10):
  if k > 2:
    break

  print(k)
```

```python
for k in range(5):
  if k == 2:
    continue

  print(k)
```

```python
k = 0
while True:
  k = k + 1

  if (k == 2):
    print("continue next")
    continue
  if k > 4:
    break

  print(k)
```

Comprehension

```python
numbers = [1, 2, 3, 4, 5]
square = []

for i in numbers:
  square.append(i**2)

print(square)
```

```python
numbers = [1, 2, 3, 4, 5]
square = [i**2 for i in numbers]
print(square)
```

```python
numbers = [1, 2, 3, 4, 5]
square = []

for i in numbers:
  if i >= 3:
    square.append(i**2)

print(square)
```

```python
numbers = [1, 2, 3, 4, 5]
square = [i**2 for i in numbers if i >= 3]
print(square)
```
